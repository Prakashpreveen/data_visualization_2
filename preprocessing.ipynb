```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel('dealer.xlsx')
data
```

```python
df = data.copy()
df.head()
```

```python
df.dtypes
```

```python
df.isna().sum()
```

```python
df_user = df.dropna(subset=['Dealers ID'])
```

```python
print('Original Number Of Rows:',len(df))
print('cleaned length of rows:',len(df_user))
```

```python
df_user.isna().sum()
```

```python
df_user.head()
```

```python
mean_value = df_user['MIMIS Current Staus.First Fund Released (Lakhs)'].mean()
mean_value
```

```python
df_user.loc[df_user['MIMIS Current Staus.First Fund Released (Lakhs)'].isnull(), 
            'MIMIS Current Staus.First Fund Released (Lakhs)'] = mean_value
```

```python
df_user.isna().sum()
```

```python
mean_value_1 = df_user['MIMIS Current Staus.Invoice Amount (Rs) 100%'].mean()
mean_value_2 = df_user['MIMIS Current Staus.Second Fund Released (Lakhs)'].mean()
mean_value_3 = df_user['Total Fund Released From TANHODA'].mean()
```

```python
df_user.loc[df_user['MIMIS Current Staus.Invoice Amount (Rs) 100%'].isnull(), 
            'MIMIS Current Staus.Invoice Amount (Rs) 100%'] = mean_value_1

df_user.loc[df_user['MIMIS Current Staus.Second Fund Released (Lakhs)'].isnull(), 
            'MIMIS Current Staus.Second Fund Released (Lakhs)'] = mean_value_2

df_user.loc[df_user['Total Fund Released From TANHODA'].isnull(), 
            'Total Fund Released From TANHODA'] = mean_value_3
```

```python
df_user.isna().sum()
```

```python
df_user.loc[df_user['MIMIS Current Staus.Second Fund Proceeding No'].isnull(),  
            'MIMIS Current Staus.Second Fund Proceeding No'] = 'Unknown'
```

```python
df_user.loc[df_user['MIMIS Current Staus.Second Fund UTR No'].isnull(),  
            'MIMIS Current Staus.Second Fund UTR No'] = 'Unknown'
```

```python
df_user.loc[df_user['MIMIS Current Staus.Current Status Remarks'].isnull(),  
            'MIMIS Current Staus.Current Status Remarks'] = 'Unknown'
```
